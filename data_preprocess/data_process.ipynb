```python
import json

data_file = "../data/WN18RR/facts.txt"
with open(data_file) as f:
    data = f.readlines()

entity_file = "../data/WN18RR/entities.json"
with open(entity_file) as f:
    entity_data = json.load(f)
```

```python
l = list()
for item in data:
    d = dict()
    head_id, relation, tail_id = item.split('\n')[0].split('\t')
    head = next(filter(lambda x: x['entity_id'] == head_id, entity_data), None)["entity"]
    tail = next(filter(lambda x: x['entity_id'] == tail_id, entity_data), None)["entity"]
    d["head_id"] = head_id
    d["head"] = head
    d["relation"] = relation
    d["tail_id"] = tail_id
    d["tail"] = tail
    l.append(d)

save_path = "../data/WN18RR/facts.txt.json"
with open(save_path, 'w') as f:
    json.dump(l, f,indent=4)
```
